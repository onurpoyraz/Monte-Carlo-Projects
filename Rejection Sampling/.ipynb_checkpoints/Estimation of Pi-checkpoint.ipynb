## Estimation of Pi by Rejection Sampling

```python
%matplotlib inline
import pandas as pd
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from ipywidgets import interact, IntSlider
```

```python
%config InlineBackend.figure_format = 'retina'
plt.rcParams['figure.dpi'] = 200
plt.style.use('dark_background')
```

By the definition of the area of the circle $A = \pi*r^2$. Which means that on the unit square the area of the unit circle is $\pi/4$. Therefore the the ratio $\frac{accepted sample}{total sample} = \frac{\pi}{4}$. So if we multiply the ratio of accepted sample over total sample by 4 we can find $\pi$.

```python
def rejection_sampling(N):
    x = np.random.uniform(-1,1,N).reshape(N,1)
    y = np.random.uniform(-1,1,N).reshape(N,1)
    xy = np.concatenate((x,y), axis = 1)
    fig = plt.figure(figsize=(10,10))
    plt.xlim(-1,1)
    plt.ylim(-1,1)
    ax = fig.add_subplot(1, 1, 1)
    circ = plt.Circle((0, 0), radius=1, edgecolor='w', facecolor='None')
    ax.add_patch(circ)
    count = 0
    for item in xy:
        if (np.square(item[0]) + np.square(item[1])) <= 1.0:
            plt.plot(item[0], item[1], 'm.')
            count += 1
        else:
            plt.plot(item[0], item[1], 'c.')
    print ("Pi is equal to {}".format(4*(count/N)))
    plt.show()
```

```python
rejection_sampling(100)
```

```python
rejection_sampling(1000)
```
